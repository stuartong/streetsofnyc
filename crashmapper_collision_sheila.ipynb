```python
### SIMILAR TO TICKET DATA
def clean_street_bert(df):
    '''
    Function cleans the street names slightly for better matching in BERT
    '''
        
    # clean white space
    df['on_street_clean'] = df['on_street_name'].str.replace(' +',' ',regex=True)
    
    # standardize some common street name patterns
    df['on_street_clean'] = df['on_street_clean'].str.replace(r"(E) ([0-9]+) (ST)",'EAST \g<2> STREET',regex=True)
    df['on_street_clean'] = df['on_street_clean'].str.replace(r"(W) ([0-9]+) (ST)",'WEST \g<2> STREET',regex=True)

    df['on_street_clean'] = df['on_street_clean'].str.replace(r" ST$",'STREET',regex=True)

    df['on_street_clean'] = df['on_street_clean'].str.replace(r" AVE$",'AVENUE',regex=True)
    
    df['on_street_clean'] = df['on_street_clean'].str.replace(r" PL$",'PLACE',regex=True)

    df['on_street_clean'] = df['on_street_clean'].str.replace(r" BLVD$",'BOULEVARD',regex=True)
    
    
    # ADD MORE BELOW HERE THAT IS DIFFERENT FROM TICKET DATA
    df['on_street_clean'] = df['on_street_clean'].str.replace(r" ROADWA$",'ROADWAY',regex=True)
    
    df['on_street_clean'] = df['on_street_clean'].str.replace(r" EXPY$",'EXPRESSWAY',regex=True)
    
    df['on_street_clean'] = df['on_street_clean'].str.replace(r" PKWY$",'PARKWAY',regex=True)
    
    
    
    ### DO IT ON ANOTHER COLUMN
    # clean white space
    df['cross_street_clean'] = df['cross_street_name'].str.replace(' +',' ',regex=True)
    
    # standardize some common street name patterns
    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r"(E) ([0-9]+) (ST)",'EAST \g<2> STREET',regex=True)
    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r"(W) ([0-9]+) (ST)",'WEST \g<2> STREET',regex=True)

    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r" ST$",'STREET',regex=True)

    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r" AVE$",'AVENUE',regex=True)
    
    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r" PL$",'PLACE',regex=True)

    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r" BLVD$",'BOULEVARD',regex=True)
    
    
    # ADD MORE BELOW HERE THAT IS DIFFERENT FROM TICKET DATA
    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r" ROADWA$",'ROADWAY',regex=True)
    
    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r" EXPY$",'EXPRESSWAY',regex=True)
    
    df['cross_street_clean'] = df['cross_street_clean'].str.replace(r" PKWY$",'PARKWAY',regex=True)
    
    return (df)
```
