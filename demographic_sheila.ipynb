# DEMOGRAPHIC

```python
path = os.getcwd()
keyword = 'demographic'

demographic = []
for fname in os.listdir('/Users/sheila/milestone1/cleaning/assets'):
    if keyword in fname:
        demographic.append('assets/'+fname)
```

```python
df = pd.DataFrame()

for i in demographic:
    data = pd.read_excel(i)
    
    data = data.iloc[2:].drop(columns=('Unnamed: 2')).T.reset_index()
    
    data.columns = data.iloc[0]
    data = data.drop(data.index[0])
    
    #FIGURE OUT HOW TO CHANGE FLOAT HEADER NAMES!
#     for col in test.columns:
#         print(col.lstrip())
    
    df = df.append(data)
    
df.head()
```
