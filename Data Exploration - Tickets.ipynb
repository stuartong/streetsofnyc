# Data Cleaning and Processing

## Step 1:  Load Data from feather

```python
ticket20 = feather.read_dataframe('ticket_2020.feather')
```

```python
ticket20.sample(10)
```

## Step 2: Clean up Violation Codes and Get Boro Code

### Update Violation Codes for clean definition of violation

```python
# get violation codes

violation_codes =pd.read_json("https://data.cityofnewyork.us/resource/ncbg-6agr.json")
violation_dict = violation_codes.set_index('code').to_dict()['definition']
ticket20['Clean Violation Des'] = ticket20['Violation Code'].map(violation_dict)
ticket20.head(5)
```

### Create Boro Code from Violation County

```python
ticket20.groupby('Violation County').count()
```

```python
# P we treat as error/none and count is boro 0 - also one data point


boro_dict = {'NY':'1',
             'MN':'1',
             'BX':'2',
             'K':'3',
             'K F':'3',
             'BK':'3',
             'KINGS':'3',
             'Q':'4',
             'P' : '0',
             'QN':'4',
             'QUEEN':'4',
             'QNS':'4',
             'R':'5',
             'ST':'5',
             'None':'0'
            }
```

```python
ticket20['Boro Code'] = ticket20['Violation County'].map(boro_dict)
ticket20.head(5)
```

## No Street Codes

Since we will intend to use street codes to ultimately join the data to coordinate/streets, let's check out tickets with ALL street codes = 0

```python
no_street_code=ticket20[(ticket20['Street Code1']==0)&(ticket20['Street Code2']==0)&(ticket20['Street Code3']==0)]
```

```python
no_street_code.sample(10)
```

4.4mn out of 12mn entries has no street code - quite problematic. Check the distribution by violation and boro code

```python
no_street_code.groupby(['Boro Code','Clean Violation Des']).agg({'Summons Number':'count'}).sort_values(by=['Summons Number'],ascending=False)
```

```python
nocodevio=no_street_code.groupby(['Clean Violation Des']).agg({'Summons Number':'count'}).sort_values(by=['Summons Number'],ascending=False)
nocodevio
```

But looks like the majority are related to school zone speed violations

```python
allvio=ticket20.groupby(['Clean Violation Des']).agg({'Summons Number':'count'}).sort_values(by=['Summons Number'],ascending=False)
allvio
```

```python
allvio.reset_index(inplace=True)
```

```python
allvio.rename(columns={'Summons Number':'Total Count'},inplace=True)
```

```python
nocodevio_dict = nocodevio.to_dict()['Summons Number']
allvio['No SC Count']=allvio['Clean Violation Des'].map(nocodevio_dict)
```

```python
totalvios = allvio['Total Count'].sum()
totalnosc = allvio['No SC Count'].sum()

allvio['proportion of all'] = allvio['Total Count']/totalvios
allvio['proportion ticket type with no sc']=allvio['No SC Count']/allvio['Total Count']
allvio['proportion of no sc']=allvio['No SC Count']/totalnosc

allvio.sort_values(by=['proportion of all'],ascending=False).head(20)
```

```python
schoolzone.groupby(['Street Name','Intersecting Street']).agg({'Summons Number':'count'}).sort_values(by=['Summons Number'],ascending=False).head(50)
```

## Conclusion

As street codes will be the main matcher - we will do the following for both datasets:
1. Match School zone speeding violations (No street codes) with speed camera locations to get coordinates
2. Match failure to stop at red light violations with camera locations to get coordinates

These 2 violation types are substanial in terms of representation and hence we will try to "salvage" the data points for our analysis. 

For the other data missing street codes -they typically represent either:
1. A small portion of the total data
2. A small portion of that violation type

Hence we choose to drop those datapoints
